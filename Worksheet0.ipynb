# Explanation of Error:
The error occurs because NumPy requires arrays to have the same number of dimensions when using concatenate, vstack, or hstack.
Here, x is 2D (shape (2,2)) while v is 1D (shape (2,)), so they cannot be combined directly.

```python
#Problem4
#Task1
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])

print("Array A:\n", A)
print("Array B:\n", B)
```

```python
#Question1
A_inv = np.linalg.inv(A)
product_AA_inv = np.dot(A, A_inv)

print("Inverse of A (A⁻¹):\n", A_inv)
print("Product A * A⁻¹:\n", product_AA_inv)
print("Is A * A⁻¹ approximately equal to Identity matrix?\n", np.allclose(product_AA_inv, np.eye(A.shape[0])))
```

# Task2
The system of linear equations is:

2x - 3y + z = -1
x - y + 2z = -3
3x + y - z = 9

This can be represented in matrix form as AX = B, where:

A = [[2, -3, 1],
     [1, -1, 2],
     [3,  1, -1]]

X = [[x],
     [y],
     [z]]

B = [[-1],
     [-3],
     [9]]

```python
# Represent the coefficients matrix A
A_coeffs = np.array([[2, -3, 1],
                     [1, -1, 2],
                     [3,  1, -1]])

# Represent the constants vector B
B_constants = np.array([[-1],
                        [-3],
                        [9]])

print("Coefficient Matrix A:\n", A_coeffs)
print("Constants Vector B:\n", B_constants)
```

```python
#Task3
# Calculate the inverse of A
A_coeffs_inv = np.linalg.inv(A_coeffs)

print("Inverse of A (A⁻¹):\n", A_coeffs_inv)

# Solve for X = A⁻¹B
X_solution = np.dot(A_coeffs_inv, B_constants)

print("Solution for X (x, y, z):\n", X_solution)

x_val = X_solution[0][0]
y_val = X_solution[1][0]
z_val = X_solution[2][0]

print(f"The solution is: x = {x_val:.2f}, y = {y_val:.2f}, z = {z_val:.2f}")
```
